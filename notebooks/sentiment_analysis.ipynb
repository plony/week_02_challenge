```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Load cleaned CSVs
df_cbe = pd.read_csv("../../data/processed/cbe_reviews_cleaned.csv")
df_boa = pd.read_csv("../../data/processed/boa_reviews_cleaned.csv")
df_dashen = pd.read_csv("../../data/processed/dashen_reviews_cleaned.csv")

# Create images directory if not exists
os.makedirs("../../assets/images", exist_ok=True)

# Plot CBE Sentiment
plt.figure(figsize=(6, 4))
sns.barplot(x=df_cbe['sentiment'].value_counts().index,
            y=df_cbe['sentiment'].value_counts().values, palette='viridis')
plt.title("CBE - Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.savefig("../../assets/images/cbe_sentiment.png")
plt.show()

# Plot BOA Sentiment
plt.figure(figsize=(6, 4))
sns.barplot(x=df_boa['sentiment'].value_counts().index,
            y=df_boa['sentiment'].value_counts().values, palette='viridis')
plt.title("BOA - Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.savefig("../../assets/images/boa_sentiment.png")
plt.show()

# Plot Dashen Sentiment
plt.figure(figsize=(6, 4))
sns.barplot(x=df_dashen['sentiment'].value_counts().index,
            y=df_dashen['sentiment'].value_counts().values, palette='viridis')
plt.title("Dashen - Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.savefig("../../assets/images/dashen_sentiment.png")
plt.show()
```
